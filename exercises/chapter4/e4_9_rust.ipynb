## Exercise 4.9
### $p_h = 0.25$

```rust
use e4_9::*;
use dashmap::DashMap;

let gamblers_env_25 = GamblersEnv::new(/*p_h=*/0.51);
let (policy_25, value_25) = gamblers_env_25.value_iteration(/*theta=*/ 1e-5);
```

#### plotting policy

```rust
use dashmap::DashMap;
use plotters::prelude::*;

fn bar(s:i32, a: i32) -> Rectangle<(i32, i32)>{
    let mut bar = Rectangle::new([(s, 0), (s + 1, a)], BLACK.filled());
    bar.set_margin(0, 0, 0, 0);
    bar
}
```

```rust
evcxr_figure((500,500), |root| {
    root.fill(&WHITE)?;
    let mut chart = ChartBuilder::on(&root)
        .caption("Gamblers Policy Funtion p_h=0.25", ("sans-serif", 30))
//        .margin(5)
        .set_label_area_size(LabelAreaPosition::Left, 100)
        .set_label_area_size(LabelAreaPosition::Bottom, 100)
        .build_ranged(0i32..100i32, 0i32..100i32)?;

    chart.configure_mesh()
    .x_desc("captital")
    .y_desc("stake")
    .draw()?;
    
    chart.draw_series(
        gamblers_env_25.states().iter().map(|s| {
            let a = *policy_25.get(s).unwrap();
            bar(*s,a)
        })
    )?;

    // Tell plotters that everything is ok
    Ok(())
}).style("width: 400px")
```

#### value function

```rust
extern crate plotters;
use plotters::prelude::*;

evcxr_figure((500,500), |root| {
    root.fill(&WHITE)?;
    let mut chart = ChartBuilder::on(&root)
        .caption("Gamblers Value Funtion p_h=0.25", ("sans-serif", 30))
//        .margin(5)
        .set_label_area_size(LabelAreaPosition::Left, 100)
        .set_label_area_size(LabelAreaPosition::Bottom, 100)
        .build_ranged(0.0..100.0, 0.0..1.0)?;

    chart.configure_mesh()
    .x_desc("captital")
    .y_desc("value")
    .draw()?;
    
    chart.draw_series(
        gamblers_env_25.states().iter().map(|s| {
            let v = *value_25.get(s).unwrap();
            let mut bar = Rectangle::new([(*s as f64, 0.0), (*s as f64 + 1.0, v)], BLUE.filled());
            bar.set_margin(0, 0, 0, 0);
            bar
        })
    )?;

    // Tell plotters that everything is ok
    Ok(())
}).style("width: 400px")
```
