```python
import future
import numpy as np
class Bandit():
    def __init__(self, h, k=10):
        self.a = None
        self.k = k

    def action(self) -> int:
        raise NotImplementedError()

    def reward(self, r_t:float) -> None:
        raise NotImplementedError()

    def get_k():
        return self.k
```

```python
class ValueExeption(BaseException):
    pass
```

```python
"""
Abstract class for value based bandits using constent step size
"""
class ValueConstStepBandit(Bandit):
    def __init__(self, eplsilon, alpha=0.1, k=10):
        Bandit.__init__(self, eplsilon, k=k)
        self.alpha = alpha
        self.Q = np.empty(k)
    
    def reward(self, r_t:float):
        self.Q[self.a] += self.alpha*(r_t - self.Q[self.a]) 

"""
Epsilon Greedy bandit with hyper perameter eplison and alpha fixed.
"""
class EpsilonGreedyBandit(ValueConstStepBandit):
    def __init__(self, epsilon, alpha=0.1, k=10):
        ValueConstStepBandit.__init__(self, 0, alpha=alpha, k=k)
        self.epsilon = epsilon
        self.Q = np.zeros(k)
        if(not 0 <= self.epsilon <= 1):
            raise ValueExeption()

    def action(self):
        if rand.uniform() < self.epsilon:
            self.a = rand.randint(0, self.k)
        else:
            self.a = np.argmax(self.Q)
        return self.a
```

```python
"""
Optimistic bandit. Gready with fixed alpha. Hyper perameter is start value.
"""
class OptimisticBandit(EpsilonGreedyBandit):
    def __init__(self, Q_0, alpha=0.1, k=10):
        EpsilonGreedyBandit.__init__(self, 0, alpha=alpha, k=k)
        self.Q = np.full(k, Q_0)
```

```python
class UCBBandit(ValueConstStepBandit):
    def __init__(self, c, alpha=0.1, k=10):
        ValueConstStepBandit.__init__(self, c, alpha=alpha, k=k)
        self.c = c
        self.t = 1
        self.N = np.ones(k)
        if(not 0 <= c):
            raise ValueExeption()

    def action(self):
        self.a = np.argmax(self.Q + self.c*np.sqrt(np.log(self.t)/self.N))
        self.t += 1
        return self.a
```

```python
from scipy.special import softmax
"""
Gradiant based bandit with hyper perameter being its learning rate alpha
"""
class GradiantBandit(Bandit):
    def __init__(self, alpha, k=10):
        Bandit.__init__(self, alpha, k=k)
        self.alpha = alpha
        # H is the prefrence
        self.H = np.zeros(k)
        self.baseline = None
        self.t = 1
        self.policy = softmax(self.H)
        if(self.alpha < 0):
            raise ValueExeption()
    def action(self):
        self.a = rand.choice(
            np.arange(self.k), 
            p=self.policy)
        return self.a

    def reward(self, r: float):
        if self.baseline is None:
            self.baseline = r
        a = self.a
        Ha_t = self.H[a]
        self.H = self.H - self.alpha*(r - self.baseline)*self.policy
        # the selected action goes in the oposite direction
        self.H[a] = Ha_t + self.alpha*(r - self.baseline)*(1-self.policy[a])
        # we use the running avg as the baseline
        self.baseline += (1/self.t)*(r - self.baseline)
        self.policy = softmax(self.H)
        self.t += 1
    
```

```python
from tqdm.notebook import tqdm
import numpy.random as rand
"""
Simulate a k armed bandit with moving target for step steps
the agent is passed in as the agent argument.
Runs the k armed bandit test bed with moving q*.
"""
def run_k_armed(agent: Bandit, k=10, steps=200000, start_avg=100000, moving=True):
    q_stars = rand.normal(loc=0, scale=1, size=k)
    avg_reward = 0
    optimial_perportion=0
    for t in tqdm(range(1, steps + 1), leave=False):
        a_t = agent.action()
        reward = rand.normal(loc=q_stars[a_t], scale=1)
        if (t > start_avg):
            avg_reward += (1/(t - start_avg))*(reward - avg_reward)
            chose_optimial = (1 if a_t == np.argmax(q_stars) else 0)
            optimial_perportion += (1/(t - start_avg))*(chose_optimial - optimial_perportion)
        agent.reward(reward)
        # Walk the targets
        if moving:
          q_stars += rand.normal(loc=0, scale=0.01, size=k)
    return avg_reward, optimial_perportion
```

```python
samples = 100
params = np.logspace(-6, 2, base=2.0, num=20)
results_r = np.empty((samples, 4, len(params)))
results_c = np.empty((samples, 4, len(params)))
bandit_names = [ 
    "Epsilon Greedy Bandit (alpha=0.1)",
    "Optimistic Bandit (alpha=0.1)",
    "UCB Bandit (alpha=0.1)",
    "Gradiant Bandit (avg basline)"
]
for k in tqdm(range(samples)):
    for i in tqdm(range(len(params)), leave=False):
        p = params[i]
        bandits = [
            EpsilonGreedyBandit(p) if 0 <= p <= 1 else None,
            OptimisticBandit(p),
            UCBBandit(p) if p >= 0 else None,
            GradiantBandit(p) if p > 0 else None
        ]
        for j in tqdm(range(len(bandits)), leave=False):
            bandit = bandits[j]
            if bandit is not None:
                results_r[k, j, i], results_c[k, j, i] = run_k_armed(bandit, steps=2000, start_avg=1, moving=False)
```

```python
import matplotlib.pyplot as plt
ranges = [
    [0,0.1],
    [0,10],
    [0,10],
    [0,10]
]
avg_results_r = np.mean(results_r, axis=0)
avg_results_c = np.mean(results_c, axis=0)
fig, (axs_reward, axs_correct) = plt.subplots(2)
fig.suptitle('Perameter study')
for i in range(4):
    indexes = np.where(np.logical_and(params>=ranges[i][0], params<=ranges[i][1]))
    axs_reward.plot(params[indexes], avg_results_r[i, indexes].flatten())
    axs_correct.plot(params[indexes], avg_results_c[i, indexes].flatten())
axs_correct.label_outer()
axs_reward.label_outer()
axs_correct.set_title("% best action")
axs_reward.set_title("Avg reward")
axs_correct.set_xscale("log", basex=2)
axs_reward.set_xscale("log", basex=2)
plt.legend(bandit_names, bbox_to_anchor=(1.04,1), loc="upper left")
plt.show()
```
